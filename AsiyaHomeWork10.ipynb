```python
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
```

```python
new_data = data.copy()
```

```python
new_data.loc[new_data['whoAmI'] == 'robot', 'amIrobot'] = int(True)
new_data.loc[new_data['whoAmI'] == 'human', 'amIrobot'] = int(False)
new_data.loc[new_data['whoAmI'] == 'human', 'amIhuman'] = int(True)
new_data.loc[new_data['whoAmI'] == 'robot', 'amIhuman'] = int(False)
```
